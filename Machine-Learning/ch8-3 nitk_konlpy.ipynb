# NLTK 자연어 처리 패키지
### 말뭉치

```python
import nltk
nltk.download("book", quiet=True)
from nltk.book import *
```

```python
nltk.corpus.gutenberg.fileids()
```

```python
emma_raw = nltk.corpus.gutenberg.raw("austen-emma.txt")
print(emma_raw[:1302])
```

### 토큰 생성

```python
# 문장단위로 나누기

from nltk.tokenize import sent_tokenize
print(sent_tokenize(emma_raw[:1000])[3])
```

```python
# 단어단위로 나누기 (구둣점 포함)

from nltk.tokenize import word_tokenize
word_tokenize(emma_raw[50:100])
```

```python
# 단어단위로 나누기 (구둣점 없음)

from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")
retokenize.tokenize(emma_raw[50:100])
```

### 형태소 분석
- 어간 추출과 원형 복원

```python
from nltk.stem import PorterStemmer, LancasterStemmer

st1 = PorterStemmer()
st2 =  LancasterStemmer()

words = ["fly", "flies", "flying", "flew", "flown"]

print("Porter Stemmer   :", [st1.stem(w) for w in words])
print("Lancaster Stemmer:", [st2.stem(w) for w in words])
```

- 원형 복원

```python
from nltk.stem import WordNetLemmatizer

lm = WordNetLemmatizer()

[lm.lemmatize(w, pos="v") for w in words]
```

### 품사 부착
- NNP: 단수 고유명사
- VB: 동사
- VBP: 동사 현재형
- TO: to 전치사
- NN: 명사(단수형 혹은 집합형)
- DT: 관형사

```python
nltk.help.upenn_tagset("VB")   # 설명이 쭉 나옴
```

```python
from nltk.tag import pos_tag 
# pos_tag 명령을 사용하면 단어 토큰에 품사를 부착하여 튜플로 출력

sentence = "Emma refused to permit us to obtain the refuse permit"
tagged_list = pos_tag(word_tokenize(sentence))
tagged_list
```

```python
# 명사 추출
nouns_list = [t[0] for t in tagged_list if t[1] == "NN"]  # 리스트 표현식 이해하기
nouns_list
```

```python
from nltk.tag import untag
# untag는 태그 튜플, 품사 태그 정보 제거하기

untag(tagged_list)
```

```python
def tokenizer(doc):
    return ["/".join(p) for p in tagged_list]  # 중간에 /를 넣고 결합시킴

tokenizer(sentence)
```

### Text 클래스

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False  # minus 표시
mpl.rc('font', family = 'Malgun Gothic')
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (13,10)
plt.rcParams["font.size"] = 14
```

```python
from nltk import Text

text = Text(retokenize.tokenize(emma_raw))
```

```python
text.plot(20)
plt.show()
```

```python
text.dispersion_plot(["Emma", "Knightley", "Frank", "Jane", "Harriet", "Robert"])
```

```python
text.concordance("Emma") #메서드로 단어가 사용된 위치를 직접 표시, 문맥파악에 도움
```

```python
text.similar("Emma") # Emmma를 대체하는 다른 단어를 예상해줌
```

```python
text.common_contexts(["Emma", "she"]) # 둘이 같이 나올 때의 문맥을 이루는 단어들?
```

### FreqDist

```python
fd = text.vocab()
type(fd)
```

```python
from nltk import FreqDist

stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"]
emma_tokens = pos_tag(retokenize.tokenize(emma_raw))
names_list = [t[0] for t in emma_tokens if t[1] == "NNP" and t[0] not in stopwords]
fd_names = FreqDist(names_list)
```

```python
fd_names.N(), fd_names["Emma"], fd_names.freq("Emma") 
# 전체에서 나오는 이름 수, 엠마가 나온 빈도, 전체 중 엠마의 빈도
```

```python
fd_names.most_common(5)
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="black", random_state=0)
plt.imshow(wc.generate_from_frequencies(fd_names))
plt.axis("off")
plt.show()
```
